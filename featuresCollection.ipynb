```python
electricityEnergySource_df = pd.read_csv("resources/electricityEnergySource_daily.csv")
electricityEnergySource_df.head()
```

```python
electricityEnergySource_df.info()
```

```python
electricityEnergySource_df['period'] = pd.to_datetime(electricityEnergySource_df['period'])
```

```python
electricityEnergySource_df.info()
```

```python
electricityEnergySource_df['type-name'].unique()
```

```python
daily_eleEnergySource_df = electricityEnergySource_df.groupby(['period', 'type-name'])['value'].sum().reset_index()
daily_eleEnergySource_df.head()
```

```python
daily_eleEnergySource_df.columns
```

```python
# Daily total electricity generation by all energy source
daily_total = daily_eleEnergySource_df.groupby('period')['value'].sum().reset_index()
daily_total.rename(columns={'value': 'ele_total'}, inplace=True)

# Daily total electricity generation by natural gas & petroleum
filtered_df = daily_eleEnergySource_df[daily_eleEnergySource_df['type-name'].isin(['Natural Gas', 'Petroleum'])]
pivot_df = filtered_df.pivot(index='period', columns='type-name', values='value').reset_index()
pivot_df.rename(columns={'Natural Gas': 'ele_ng', 'Petroleum': 'ele_oil'}, inplace=True)

# Merged Dataframe
eleEnergySource_df = pd.merge(daily_total, pivot_df, on='period', how='inner')

eleEnergySource_df.head()
```

```python
eleEnergySource_df.info()
```

```python
electricityDemandInter_df = pd.read_csv("resources/electricity_demandInterchange_daily.csv")
electricityDemandInter_df.head()
```

```python
electricityDemandInter_df.info()
```

```python
electricityDemandInter_df['period'] = pd.to_datetime(electricityDemandInter_df['period'])
```

```python
electricityDemandInter_df.info()
```

```python
electricityDemandInter_df["type-name"].unique()
```

```python
filtered_df2 = electricityDemandInter_df.groupby(['period', 'type-name'])['value'].sum().reset_index()
pivot_df2 = filtered_df2.pivot(index='period', columns='type-name', values='value').reset_index()

pivot_df2.rename(columns={'Demand': 'ele_demand', 'Total interchange': 'ele_ic'}, inplace=True)

electricity_daily = pd.merge(eleEnergySource_df, pivot_df2, on='period', how='inner').rename(columns={'period': 'Date'})

electricity_daily
```

```python
electricity_daily.info()
```

```python
model_daily = pd.merge(merged2, electricity_daily, on='Date', how='inner')
model_daily
```

```python
model_daily.to_csv("outputs/model_features_daily.csv", index=False)
```
