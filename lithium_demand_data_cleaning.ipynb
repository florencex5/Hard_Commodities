```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("resources/lithium_demand.csv")
data.head()
```

```python
data.columns = data.iloc[0]
data = data[1:]
```

```python
df_transposed = data.transpose()
df_transposed.head()
```

```python
df_transposed.columns = df_transposed.iloc[0]
df_transposed = df_transposed[1:]
df_transposed.head()
```

```python
df_cleaned = df_transposed.dropna(how='all').dropna(axis=1, how='all')
df_cleaned.head()
```

```python
scenarios = np.repeat(['STEPS', 'APS', 'NZE'],(len(df_cleaned)/ 3))
scenarios = np.insert(scenarios, 0, "BASE")

df_new = df_cleaned.assign(scenarios=scenarios)
df_new.head()
```

```python
df_new['Critical Minerals'] = "Lithium"
df_new.head()
```

```python
df_new.reset_index(drop=False, inplace=True)
df_new.head()
```

```python
df_new.rename(columns={0: 'Year'}, inplace=True)
df_new.head()
```

```python
df_new.to_excel("outputs/lithium_demand.xlsx", index=False)
```
